# Dropout Regularization in Deep Neural Network

This is a dataset that describes sonar chirp returns bouncing off different services. The 60 input variables are the strength of the returns at different angles. It is a binary classification problem that requires a model to differentiate rocks from metal cylinders

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('sonar.csv', header=None)
df.sample(5)
```

```python
# check for nan values
df.isna().sum()
```

```python
df[60].value_counts
```

```python
X = df.drop(60, axis=1)
y = df[60]
```

```python
X
```

```python
y
```

```python
y.dtypes
```

```python
y = pd.get_dummies(y, drop_first=True)
y.sample(5)     # R -->  1 & M --> 0
```

```python
y = y.astype(int)
```

```python
y.sample(5)
```

```python
y.value_counts()
```

Training & Testing Splliting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.shape , X_test.shape
```

Now Doing Neutral Network

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_shape=(60,), activation='relu'),
    keras.layers.Dense(40, activation='relu'),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer = 'adam',
              loss='binary_crossentropy',
              metrics = ['accuracy']
              )

model.fit(X_train, y_train, epochs=100, batch_size = 8 )
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# Round the values to nearest integer i.e; 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```

## Now we drop the layer.

```python
modeld = keras.Sequential([
    keras.layers.Dense(60, input_shape=(60,), activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(40, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

modeld.compile(optimizer='adam',
               loss='binary_crossentropy',
               metrics=['accuracy'])

modeld.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
modeld.evaluate(X_test, y_test)
```

```python
y_pred = modeld.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)

from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```
